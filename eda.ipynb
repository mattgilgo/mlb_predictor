```python
from pybaseball.team_batting import team_batting_bref
from pybaseball.teamid_lookup import team_ids

team_id_df = team_ids()
mlb_id_df = team_id_df[team_id_df['yearID'] == 2021]
```

```python
year = 2024
league_team_batting_stats = pd.DataFrame()
for team in mlb_id_df['teamIDBR']:
    team_stats = team_batting_bref(team, year)
    league_team_batting_stats = pd.concat([league_team_batting_stats, team_stats])
```
